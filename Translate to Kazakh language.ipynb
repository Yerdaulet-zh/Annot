```python
import json, time, tqdm
from googletrans import Translator

translator = Translator()

def preprocess(translated):
    translated = translated.replace("\'s", "'s")
    translated = translated.replace('"', "'")
    translated = translated.replace(")", ") ")
    translated = translated.replace("(", " (")
    translated = translated.replace(".", ". ")
    translated = translated.replace(",", ", ")
    translated = translated.replace("  ", " ")
    translated = translated.replace("  ", " ")
    return translated.strip()
```

```python
Contexts = []
prev_context = ""

retry = 0

for context in tqdm.tqdm(contexts):
    if context != prev_context:
        prev_context = context
        context = preprocess(context)
        while retry != 5:
            try:
                translated = translator.translate(context, 'kk').text
                Contexts.append(translated)
                break
            except Exception as e:
                retry += 1
                print(e)
                translated = "pass"
        else:
            retry = 0
            Contexts.append(translated)
    else:
        Contexts.append(translated)
```

```python
Questions = []
prev_context = ""

retry = 0

for context in tqdm.tqdm(questions):
    if context != prev_context:
        prev_context = context
        context = preprocess(context)
        while retry != 5:
            try:
                translated = translator.translate(context, 'kk').text
                Questions.append(translated)
                break
            except Exception as e:
                retry += 1
                print(e)
                translated = "pass"
        else:
            retry = 0
            Questions.append(translated)
    else:
        Questions.append(translated)
```

```python
Answers = []
prev_context = ""

retry = 0

for context in tqdm.tqdm(answers):
    context = context['text'][0]
    if context != prev_context:
        prev_context = context
        context = preprocess(context)
        while retry != 5:
            try:
                translated = translator.translate(context, 'kk').text
                Answers.append(translated)
                break
            except Exception as e:
                retry += 1
                print(e)
                translated = "pass"
        else:
            retry = 0
            Answers.append(translated)
    else:
        Answers.append(translated)
```

```python
import json
from googletrans import Translator
import time

translator = Translator()


def translate_data(data, target_language, max_retries=5):
    if isinstance(data, dict):
        return {key: translate_data(value, target_language) for key, value in data.items()}
    elif isinstance(data, list):
        return [translate_data(value, target_language) for value in data]
    elif isinstance(data, str):
        if data:
            retries = 0
            while retries < max_retries:
                try:
                    return translator.translate(data, dest=target_language).text
                except Exception as e:
                    print(f"Translation error: {e}. Retrying...")
                    retries += 1
                    time.sleep(1)  # You might want to wait a bit before retrying
            print(f"Failed to translate text after {max_retries} retries.")
            return data
        else:
            return data
    else:
        return data

# The rest of your functions (translate_batch, translate_and_update_json) remain unchanged



def translate_batch(data, target_language, start, batch_size):
    end = min(start + batch_size, len(data))
    for i in range(start, end):
        data[i] = translate_data(data[i], target_language)
        print(f"Translated sample {i + 1}")
    return data


def translate_and_update_json(file_path, target_language, start_index=0):
    try:
        with open(file_path, 'r+', encoding='utf-8') as file:
            data = json.load(file)

            for i in range(start_index, len(data)):
                data[i] = translate_data(data[i], target_language)
                print(f"Translated sample {i + 1}")

                # Move the file pointer to the beginning of the file
                file.seek(0)

                # Update the JSON file with the translated data
                json.dump(data, file, ensure_ascii=False, indent=4)

                # Truncate the file to remove any remaining parts of the previous content
                file.truncate()

                # Delay between each translation to prevent rate-limiting issues
                time.sleep(0.5)

    except Exception as e:
        print(f"An error occurred: {e}")

# Usage example
#translate_and_update_json('your_json_file.json', 'es', start_index=599, batch_size=50)  # Starting from the 600th sample

# Usage example
translate_and_update_json('part_1_test.json', 'kk',501)  # Translating to Spanish in batches of 50
```
